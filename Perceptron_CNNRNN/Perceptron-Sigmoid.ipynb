## Activation Function(Sigmoid) 포함 신경망 학습

**Wt+1=Wt −𝛼∇𝐿(𝑊𝑡) 공식을 이용해서 weight와 bias 학습**

![2 input, 1 output 학습](And_Gate.png)

```python
    X = [
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]
    ]
    Y = [0, 0, 0, 1]
```

```python
import random
import math
weights = [random.uniform(-1, 1) for _ in range(2)]
bias = random.uniform(-1, 1)
learning_rate = 0.01
epochs=5000
```

```python
def sigmoid(x):
    """시그모이드 활성화 함수"""
    try:
        return 1 / (1 + math.exp(-x))
    except OverflowError:
        return 0 if x < 0 else 1

def sigmoid_derivative(x):
    """시그모이드 함수의 미분값 (경사하강법 계산용)"""
    s = sigmoid(x)
    return s * (1 - s)
    
def predict(inputs):
    """출력 예측 (순전파)"""
    weighted_sum = sum(w * i for w, i in zip(weights, inputs)) + bias
    return sigmoid(weighted_sum)
```

```python
for epoch in range(epochs):
    total_error=0
    for inputs,label in zip(X,Y):
        #weighted_sum = sum(w * i for w, i in zip(weights, inputs)) + bias
        prediction = predict(inputs)
        error=prediction-label
        total_error += error**2
        # 3. 경사 계산 및 가중치 업데이트 (경사하강법)
        # d(Error)/d(Weight) = (prediction - label) * sigmoid'(weighted_sum) * input
        # d(Error)/d(Bias) = (prediction - label) * sigmoid'(weighted_sum)
        delta = error 
        for i in range(len(weights)):
            weights[i] -= learning_rate * delta * inputs[i]
        bias -= learning_rate * delta
        # 1000회마다 오차 출력
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}: Loss = {total_error / len(X):.4f}")
        print(weights, bias)
```

```python
print("\n--- 학습 결과 테스트 ---")
for inputs in X:
    result = predict(inputs)
    print(f"입력: {inputs}, 예측값: {result:.4f} -> 결과: {1 if result >= 0.5 else 0}")
```
